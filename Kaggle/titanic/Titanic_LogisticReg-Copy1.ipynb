```python
#必要ライブラリのインポート
import pandas as pd
from sklearn.linear_model import LogisticRegression
```

```python
#データのロード
train = pd.read_csv("train.csv")
```

```python
#データを分割
train_x = train.drop(["Name","Survived","Ticket","Cabin","Embarked"], axis=1)
train_y = train["Survived"]

#欠損値の確認
train_x.isnull().sum()
```

```python
train_x["Sex"] = train_x["Sex"].str.replace('female','2')
train_x["Sex"] = train_x["Sex"].str.replace('male','1')
train_x["Sex"]
```

```python
average_age = train_x["Age"].mean()

#欠損値を平均年齢で穴埋め
train_x["Age"] = train_x["Age"].fillna(average_age)
train_x["Age"]

train_x.isnull().sum()
```

```python
#モデルの構築
model = LogisticRegression(C=15, random_state=1)
```

```python
#学習
model.fit(train_x, train_y)
```

```python
#テスト
print(model.score(train_x, train_y))
```
