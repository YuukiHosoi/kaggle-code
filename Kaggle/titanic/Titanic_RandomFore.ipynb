```python
#必要ライブラリのインポート
import pandas as pd
import csv as csv
from sklearn.ensemble import RandomForestClassifier
```

```python
#データのロード
train = pd.read_csv("train.csv")
```

```python
#データを分割
train_x = train.drop(["Name","Survived","Ticket","Cabin","Embarked",], axis=1)
train_y = train["Survived"]

#欠損値の確認
train_x.isnull().sum()
```

```python
train_x["Sex"] = train_x["Sex"].str.replace('female','2')
train_x["Sex"] = train_x["Sex"].str.replace('male','1')
train_x["Sex"]
```

```python
average_age = train_x["Age"].mean()

#欠損値を平均年齢で穴埋め
train_x["Age"] = train_x["Age"].fillna(average_age)
train_x["Age"]

train_x.isnull().sum()
```

```python
#モデルの構築
# for i in range(1,100):
#     model = RandomForestClassifier(n_estimators=i, max_depth=None, max_features='auto', 
#                                    n_jobs=None, random_state=0,verbose=0)
#     model.fit(train_x, train_y)
#     print(str(i) + ":" + str(model.score(train_x, train_y)))
```

```python
model = RandomForestClassifier(n_estimators=17, max_depth=5, max_features='auto', 
                                   n_jobs=None, random_state=0,verbose=0)
model.fit(train_x, train_y)
print(model.score(train_x, train_y))
```

```python
test = pd.read_csv("test.csv")
test
```

```python
ids = test["PassengerId"].values
test = test.drop(["Name","Ticket","Cabin","Embarked"], axis=1)
test
```

```python
train_x.isnull().sum()
```

```python
#欠損値を平均年齢で穴埋め
test["Age"] = test["Age"].fillna(average_age)
test["Fare"] = test["Fare"].fillna(test["Fare"].mean())

test.isnull().sum()
```

```python
test["Sex"] = test["Sex"].str.replace('female','2')
test["Sex"] = test["Sex"].str.replace('male','1')
test["Sex"]
```

```python
result = model.predict(test)
```

```python
submit_file = open("titanic_submit2.csv", "w")
file_object = csv.writer(submit_file)
file_object.writerow(["PassengerId", "Survived"])
file_object.writerows(zip(ids, result))
submit_file.close()
```
